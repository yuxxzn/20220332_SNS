```python
from bs4 import BeautifulSoup
from selenium import webdriver
import urllib.request
import time
import os
import re

print("=" * 70)
print('pixabay 사이트에서 이미지를 검색하여 수집하는 크롤러 입니다')
print("=" * 70)

keyword = input('검색어 : ')
file_no = int(input('다운로드 시도할 최대 이미지 수 : '))
f_dir = input('파일이 저장될 경로만 쓰세요 (예: c:\data\) : ')
query_txt = '사진저장'

def clean_filename(filename):
    # 유효하지 않은 문자를 공백으로 대체
    cleaned_filename = re.sub(r'[\\/:"*?<>|]', '', filename)
    return cleaned_filename
    
# 파일을 저장할 폴더를 생성
now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

os.makedirs(os.path.join(f_dir, s + '-' + query_txt), exist_ok=True)
save_path = os.path.join(f_dir, s + '-' + query_txt)

# 크롬 드라이버를 사용해서 웹 브라우저를 실행
path = "C:\\Temp\\chromedriver_win32\\chromedriver.exe"
driver = webdriver.Chrome()

s_time = time.time()  # 크롤링 시작 시간을 위한 타임 스탬프

driver.get('https://pixabay.com/images/search/' + keyword)
time.sleep(2)  


time.sleep(2)

# 스크롤을 페이지의 맨 아래로 이동
def scroll_down(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)

scroll_down(driver)

# 본문의 사진 정보를 가져와서 지정된 폴더에 저장
count = 0
img_src2 = []

while count < file_no:
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    img_elements = soup.select('div.container--wYO8e img')
    
    for img in img_elements:
        srcset = img.get('srcset')
        if srcset:
            src = srcset.split(',')[0].split()[0]
        else:
            src = img['src']
        
        img_src2.append(src)
        count += 1
        
        if count >= file_no:
            break

    if count < file_no:
        scroll_down(driver)
    else:
        break

for idx, img_url in enumerate(img_src2):
    try:
        img_path = os.path.join(save_path, f"{idx+1}.jpg")  # 파일 이름을 .jpg로 변경
        urllib.request.urlretrieve(img_url, img_path)
        print(f"{idx+1} 번째 이미지 저장중입니다=======")
    except Exception as e:
        print(f"이미지 저장 실패: {e}")
        continue
        
# 요약 정보를 출력
e_time = time.time()
t_time = e_time - s_time

store_cnt = len(img_src2)

print("=" * 70)
print("총 소요시간은 %s 초 입니다 " % round(t_time, 1))
print("총 저장 건수는 %s 건 입니다 " % store_cnt)
print("파일 저장 경로: %s 입니다" % save_path)
print("=" * 70)

driver.close()
```
